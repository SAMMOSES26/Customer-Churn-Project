# Data Generation

This notebook generates a synthetic customer churn dataset for a UK-based SaaS company.

The generated dataset includes customer demographics, subscription details, product usage metrics, customer support interactions, engagement indicators, revenue metrics, and churn information.

# Step 1: Basic Dataset

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n_rows = 15000

df = pd.DataFrame()

df["customer_id"] = [f"CF{i:05d}" for i in range(1, n_rows + 1)]

df["region"] = np.random.choice(
    [
        "London",
        "South East",
        "South West",
        "East Midlands",
        "West Midlands",
        "North East",
        "North West",
        "Scotland",
        "Wales"
    ],
    n_rows
)

df["subscription_plan"] = np.random.choice(
    ["Starter", "Professional", "Enterprise"],
    n_rows,
    p=[0.5, 0.35, 0.15]
)

df["contract_type"] = np.random.choice(
    ["Monthly", "Annual"],
    n_rows,
    p=[0.7, 0.3]
)

df["tenure_months"] = np.random.exponential(
    scale=18,
    size=n_rows
).astype(int) + 1

df["monthly_logins"] = np.random.poisson(
    lam=45,
    size=n_rows
)

df["customer_satisfaction_score"] = np.random.randint(
    1,
    11,
    n_rows
)

# Industry
df["industry"] = np.random.choice(
    [
        "Retail",
        "Finance",
        "Healthcare",
        "Education",
        "Logistics",
        "Technology",
        "Manufacturing"
    ],
    n_rows
)

# Company Size
df["company_size"] = np.random.choice(
    [
        "Micro",
        "Small",
        "Medium",
        "Large"
    ],
    n_rows,
    p=[0.4, 0.35, 0.2, 0.05]
)



# Years in Business
df["years_in_business"] = np.random.randint(
    1,
    31,
    n_rows
)

# Monthly Subscription Fee
df["monthly_subscription_fee_gbp"] = np.random.choice(
    [29, 79, 199],
    n_rows,
    p=[0.5, 0.35, 0.15]
)

# Discount Percentage
df["discount_percentage"] = np.random.randint(
    0,
    41,
    n_rows
)

# Employee Count

employee_count = []

for size in df["company_size"]:

    if size == "Micro":
        employee_count.append(
            np.random.randint(1,11)
        )

    elif size == "Small":
        employee_count.append(
            np.random.randint(11,51)
        )

    elif size == "Medium":
        employee_count.append(
            np.random.randint(51,251)
        )

    else:
        employee_count.append(
            np.random.randint(251,1001)
        )

df["employee_count"] = employee_count

# Active Users
df["active_users"] = (
    df["employee_count"]
    * np.random.uniform(0.2, 0.8, n_rows)
).astype(int)

# Feature Usage Score
df["feature_usage_score"] = np.random.randint(
    1,
    101,
    n_rows
)

# Average Session Duration
df["avg_session_duration_minutes"] = np.random.randint(
    5,
    180,
    n_rows
)

# Projects Created
df["projects_created"] = np.random.randint(
    0,
    200,
    n_rows
)

# Integrations Enabled
df["integrations_enabled"] = np.random.randint(
    0,
    20,
    n_rows
)

# Support Tickets
df["support_tickets_last_6m"] = np.random.poisson(
    2,
    n_rows
)

# Resolution Time
df["avg_resolution_time_hours"] = np.random.randint(
    1,
    72,
    n_rows
)

# Payment Failures
df["payment_failures_last_12m"] = np.random.randint(
    0,
    6,
    n_rows
)

# Email Open Rate
df["email_open_rate"] = np.random.randint(
    0,
    101,
    n_rows
)

df.loc[
    df["subscription_plan"] == "Starter",
    "email_open_rate"
] = np.where(
    np.random.rand(
        len(df[df["subscription_plan"] == "Starter"])
    ) < 0.08,
    np.nan,
    df.loc[
        df["subscription_plan"] == "Starter",
        "email_open_rate"
    ]
)

# Webinar Attendance
df["webinar_attendance_count"] = np.random.randint(
    0,
    12,
    n_rows
)

# Monthly Recurring Revenue
df["monthly_recurring_revenue"] = (
    df["monthly_subscription_fee_gbp"]
    * np.random.uniform(0.8, 1.2, n_rows)
).round(2)

# Lifetime Value
df["lifetime_value"] = (
    df["monthly_recurring_revenue"]
    * df["tenure_months"]
).round(2)

churn_score = (
    (df["tenure_months"] < 6).astype(int)
    + (df["monthly_logins"] < 20).astype(int)
    + (df["customer_satisfaction_score"] < 5).astype(int)
    + (df["payment_failures_last_12m"] > 2).astype(int)
    + (df["contract_type"] == "Monthly").astype(int)
    + (df["feature_usage_score"] < 30).astype(int)
    + (df["support_tickets_last_6m"] > 5).astype(int)
)

df["signup_date"] = pd.to_datetime(
    np.random.choice(
        pd.date_range("2020-01-01", "2025-06-01"),
        len(df)
    )
)

df["last_login_date"] = (
    df["signup_date"]
    + pd.to_timedelta(
        np.random.randint(
            1,
            365,
            len(df)
        ),
        unit="D"
    )
)

today = pd.Timestamp("2025-06-20")

df["last_login_date"] = np.where(
    df["last_login_date"] > today,
    today,
    df["last_login_date"]
)

df["last_login_date"] = pd.to_datetime(
    df["last_login_date"]
)

df["customer_segment"] = np.random.choice(
    ["SMB", "Mid-Market", "Enterprise"],
    len(df),
    p=[0.6, 0.3, 0.1]
)

df["acquisition_channel"] = np.random.choice(
    [
        "Organic Search",
        "Google Ads",
        "LinkedIn",
        "Referral",
        "Direct",
        "Partner"
    ],
    len(df)
)

df["churn"] = np.where(
    churn_score >= 3,
    "Yes",
    "No"
)
df.head()
```

# Imbalance the dataset by randomly converting some churned customers to retained

```python
yes_idx = df[df["churn"] == "Yes"].index

# randomly convert 1500 churned customers to retained
change_idx = np.random.choice(
    yes_idx,
    1500,
    replace=False
)

df.loc[change_idx, "churn"] = "No"
```

# Step 2: Introduce Missing Values

```python
cols_with_missing = [
    "monthly_logins",
    "feature_usage_score",
    "customer_satisfaction_score",
    "email_open_rate",
    "avg_session_duration_minutes"
]

for col in cols_with_missing:

    missing_idx = np.random.choice(
        df.index,
        size=int(len(df) * 0.05),
        replace=False
    )

    df.loc[missing_idx, col] = np.nan
```

```python
df.isnull().sum()
```

# Step 3: Duplicates

```python
duplicates = df.sample(
    frac=0.02,
    random_state=42
)

df = pd.concat(
    [df, duplicates],
    ignore_index=True
)
```

```python
print(df.shape)
print(df.duplicated().sum())
```

# Step 4: Add Inconsistent Categories

```python
# London variations
idx = df[df["region"] == "London"].sample(
    300,
    random_state=42
).index

df.loc[idx[:100], "region"] = "london"
df.loc[idx[100:200], "region"] = "LONDON"
df.loc[idx[200:], "region"] = " London "


# Scotland variations
idx = df[df["region"] == "Scotland"].sample(
    150,
    random_state=43
).index

df.loc[idx[:50], "region"] = "scotland"
df.loc[idx[50:100], "region"] = "SCOTLAND"
df.loc[idx[100:], "region"] = " Scotland "


# Wales variations
idx = df[df["region"] == "Wales"].sample(
    150,
    random_state=44
).index

df.loc[idx[:50], "region"] = "wales"
df.loc[idx[50:100], "region"] = "WALES"
df.loc[idx[100:], "region"] = " Wales "


# South East variations
idx = df[df["region"] == "South East"].sample(
    150,
    random_state=45
).index

df.loc[idx[:50], "region"] = "south east"
df.loc[idx[50:100], "region"] = "SOUTH EAST"
df.loc[idx[100:], "region"] = " South East "
```

# Step 5: Outliers

```python
outlier_idx = np.random.choice(
    df.index,
    150,
    replace=False
)

df.loc[
    outlier_idx,
    "monthly_logins"
] = np.random.randint(
    1000,
    5000,
    150
)
```
